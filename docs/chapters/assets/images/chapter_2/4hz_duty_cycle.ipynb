```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 1000, endpoint=False)
freq = 4 #Hz

plt.subplot(3, 1, 1)
plt.plot(t, 0.5*signal.square(2*np.pi * freq * t, duty=0.25)+0.5)
plt.grid()
plt.title("4Hz, duty cycle = 25%")
plt.xlabel("time [s]")
plt.ylabel("Signal [boolean]")

plt.subplot(3, 1, 2)
plt.plot(t, 0.5*signal.square(2*np.pi * freq * t)+0.5)

plt.grid()
plt.title("4Hz, duty cycle = 50%")
plt.xlabel("time [s]")
plt.ylabel("Signal [boolean]")

plt.subplot(3, 1, 3)
plt.plot(t, 0.5*signal.square(2*np.pi * freq * t, duty=0.75)+0.5)

plt.grid()
plt.title("4Hz, duty cycle = 75%")
plt.xlabel("time [s]")
plt.ylabel("Signal [boolean]")

plt.ylim(-0.25, 1.25)
plt.tight_layout()

plt.savefig("4hz_duty_cycle.jpg")
```
